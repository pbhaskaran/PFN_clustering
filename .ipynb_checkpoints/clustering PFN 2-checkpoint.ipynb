```python
from sklearn.datasets import make_blobs

x, y = make_blobs(n_samples=[50,50], centers=None, n_features = 2,cluster_std=1)
print(x.shape, y.shape)
plt.scatter(x[:, 0] , x[:, 1], c = y)
plt.show()
```

```python
def sample_clusters(batch_size=100, seq_len = 100, num_features =2,type_='make_blobs'):
    clusters_x = []
    clusters_y = []
    for i in range(batch_size):
        std =random.random() 
        centers = random.randint(2, 5)
        x, y = make_blobs(n_samples=seq_len, n_features=num_features,centers=centers, cluster_std=std, shuffle=True)
        clusters_x.append(x)
        clusters_y.append(y)

    clusters_x = torch.tensor(clusters_x)  
    clusters_y = torch.tensor(clusters_y)  
    clusters_x = clusters_x.permute(1, 0, 2) 
    clusters_y = clusters_y.permute(1, 0)  
    
    return clusters_x, clusters_y
```

```python
X, y = sample_clusters(batch_size=10, seq_len = 100, num_features =2,type_='make_blobs')
%matplotlib inline

# Set up grid for subplots
fig, axes = plt.subplots(3, 3, figsize=(15, 8)) 
axes = axes.flatten()  

for batch_index in range(9):
    ax = axes[batch_index] 
    ax.scatter(X[:, batch_index, 0].numpy(), X[: , batch_index, 1].numpy(), c = y[:, batch_index].numpy())
    ax.set_title(f"Batch {batch_index + 1}")

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()


```
